```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10

# 匯入ImageDataGenerator 這個類別，此類別功能強大，在此用它來對圖片做標準化
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

顯示前10張圖片(標準化前)

```python
for i in range(10):
    plt.subplot(2,5, i+1)
    plt.imshow(X_train[i])
plt.suptitle('The original image', fontsize = 12)
plt.show()
```

```python
# 建立 ImageDataGenerator的操作物件
# 將圖片各顏色通道的像素值轉換為平均值 = 0 
# 將圖片各顏色通道的像素值轉換為標準差 = 1
datagen = ImageDataGenerator(samplewise_center = True, samplewise_std_normalization = True)
```

進行標準化

```python
g = datagen.flow(X_train, y_train, shuffle = False)
X_batch, y_batch = g.next()
```

讓生成的圖片效果看起來更明顯

```python
X_batch *= 127.0 / max(abs(X_batch.min()), X_batch.max())
X_batch += 127.0
X_batch = X_batch.astype('uint8')
```

顯示前10張圖片(標準化後)

```python
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_batch[i])
    
```

```python
plt.suptitle('Standrdization result', fontsize=12)
plt.show()
```
